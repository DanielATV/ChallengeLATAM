## 2. Genera las columnas adicionales y luego expórtelas en un archivo synthetic_features.csv:

* temporada_alta : 1 si Fecha-I está entre 15-Dic y 3-Mar, o 15-Jul y 31-Jul, o 11-Sep y 30-Sep, 0 si no.

* dif_min : diferencia en minutos entre Fecha-O y Fecha-I .
* atraso_15 : 1 si dif_min > 15, 0 si no.
* periodo_dia : mañana (entre 5:00 y 11:59), tarde (entre 12:00 y 18:59) y noche (entre 19:00 y 4:59), en base a Fecha-I .

```python
#diferencia en minutos
df['dif_min'] = ((pd.to_datetime(df['Fecha-O']) - pd.to_datetime(df['Fecha-I'])).dt.total_seconds() / 60)
```

```python
#atraso de 15min
df['atraso_15'] = np.where(df['dif_min'] > 15,1,0)
```

```python
df['Fecha-I']=pd.to_datetime(df['Fecha-I'])
```

```python
#periodo_dia

def get_periodo_dia(fecha):
    
    if time(5, 0, 0)<= fecha.time() <time(12, 0, 0):
        return "mañana"
    elif time(12, 0, 0)<= fecha.time() <time(19, 0, 0):
        return "tarde"
    elif (time(19, 0, 0)<= fecha.time() <=time(23, 59, 0) or (time(0, 0, 0)<= fecha.time() <time(5, 0, 0))):
        return "noche"
    
df['periodo_dia'] = df['Fecha-I'].apply(get_periodo_dia)
```

```python
#temporada_alta
def get_temporada_alta(fecha):
    anio= int(fecha.year)
    if ((datetime.datetime(anio, 12, 15).date() <= fecha.date() <= datetime.datetime(anio, 12, 31).date())or(datetime.datetime(anio, 1, 1).date() <= fecha.date() <= datetime.datetime(anio, 3, 15).date()) or(datetime.datetime(anio, 7, 15).date() <= fecha.date() <= datetime.datetime(anio, 7, 31).date())or(datetime.datetime(anio, 9, 11).date() <= fecha.date() <= datetime.datetime(anio, 9, 30).date())):
        return 1
    else:
        return 0
df['temporada_alta'] = df['Fecha-I'].apply(get_temporada_alta)
```

```python
df[['temporada_alta','dif_min','atraso_15','periodo_dia']].to_csv('synthetic_features.csv', index = False)
```

## 3. ¿Cómo se compone la tasa de atraso por destino, aerolínea, mes del año, día de la semana, temporada, tipo de vuelo? ¿Qué variables esperarías que más influyeran en predecir atrasos?

Para calcular la tasa de atraso para cada una de las variables en primer lugar se agrupan para luego graficarlas en un gráfico de barras

```python
#destino
dfDest= df[['SIGLADES','atraso_15']]

# agrupacion
grouped_df = df.groupby('SIGLADES')['atraso_15'].agg(['count', 'mean']).reset_index()

# cambio de nombre de las columnas
grouped_df = grouped_df.rename(columns={'count': 'total', 'mean': 'ratio'})

print(grouped_df)
```

```python
#grafico
a4_dims = (11.7, 8.27)
fig, ax = plt.subplots(figsize=a4_dims)
sns.barplot(ax=ax,x='SIGLADES', y='ratio', data=grouped_df)
plt.xticks(rotation=90)
plt.show()
```

En el gráfico se ve que algunos destinos tienen una tasa muy alta en comparación al resto, sin embargo no se puede concluir al respecto puesto que, como se vio en la pregunta 1, no todos los destinos tienen la misma frecuencia, por lo que la comparación entre estos estaría sesgada.

```python
#aerolinea
dfDest= df[['OPERA','atraso_15']]

# agrupacion
grouped_df = df.groupby('OPERA')['atraso_15'].agg(['count', 'mean']).reset_index()

# cambio de nombre de las columnas
grouped_df = grouped_df.rename(columns={'count': 'total', 'mean': 'ratio'})

print(grouped_df)
```

```python
# grafico
a4_dims = (11.7, 8.27)
fig, ax = plt.subplots(figsize=a4_dims)
sns.barplot(ax=ax,x='OPERA', y='ratio', data=grouped_df)
plt.xticks(rotation=90)
plt.show()
```

Al igual que en el gráfico anterior es difícil sacar conclusiones del gráfico por sí solo, puesto que el número de vuelos por aerolínea varía significativamente. Sin embargo, se ve que hay una gran variación en la tasa de retraso dependiendo de la aerolínea.
