```python
#librerias
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## ¿Cómo se distribuyen los datos? ¿Qué te llama la atención o cuál es tu conclusión sobre esto?

```python
#data input
df = pd.read_csv('dataset_SCL.csv')
df.shape
```

Para visualizar las distribuciones de los datos se grafica un boxplot para cada variable númerica y un histograma para cada variable categorica.

```python
df.columns
```

```python
#Numero de vuelo programado
df['Vlo-I'].value_counts()    
```

```python
a4_dims = (11.7, 8.27)
fig, ax = plt.subplots(figsize=a4_dims)
sns.histplot(ax=ax,data= df['Vlo-I'].astype(str))
plt.show()
```

Graficamente se ve hay números de vuelo con una frecuencia significativamente mayor a otros.

```python
#codigo de ciudad de origen
df['Ori-I'].value_counts()
```

En este caso como todos los vuelos parten del mismo lugar, no tiene sentido analizarlo visualmente.

```python
#codigo de ciudad de origen
df['Des-I'].value_counts()
```

```python
a4_dims = (11.7, 8.27)
fig, ax = plt.subplots(figsize=a4_dims)
sns.histplot(ax=ax,data= df['Des-I'].astype(str))
plt.show()
```

En el gráfico se ve que hay destinos que tienen una frecuencia muy superior en comparación con el resto.

```python
#codigo aerolinea de vuelo
df['Emp-I'].value_counts()
```

```python
a4_dims = (11.7, 8.27)
fig, ax = plt.subplots(figsize=a4_dims)
sns.histplot(ax=ax,data= df['Emp-I'].astype(str))
plt.show()
```

En el gráfico se ve que la distrubucion de las aerolineas estan concentradas principalmente en 2 de ellas LAN y SKU.

```python
#numero de vuelo de operacion de vuelo
df['Vlo-O'].value_counts()
```

```python
a4_dims = (11.7, 8.27)
fig, ax = plt.subplots(figsize=a4_dims)
sns.histplot(ax=ax,data= df['Vlo-O'].astype(str))
plt.show()
```

En el gráfico, al igual que con el numero de vuelo programado, se ve que la frecuencia de algunos de los codigos es significativamente mayor al resto.

```python
#codigo ciudad de origen de operacion
df['Ori-O'].value_counts()
```

Puesto que todos los vuelos tienen el mismo origen no tiene sentido analizar la distrubicion de esta variable.

```python
#codigo de ciudad de destino de operación.
df['Des-O'].value_counts()
```

```python
a4_dims = (11.7, 8.27)
fig, ax = plt.subplots(figsize=a4_dims)
sns.histplot(ax=ax,data= df['Des-O'].astype(str))
plt.show()
```

En la distruicion de los codigos de las ciudades de destino se ve que los valores estan mas distrubidos, pero aun hay variables con una frecuencia muy baja.

```python
#codigo de ciudad de destino de operación.
df['Emp-O'].value_counts()
```

```python
a4_dims = (11.7, 8.27)
fig, ax = plt.subplots(figsize=a4_dims)
sns.histplot(ax=ax,data= df['Emp-O'].astype(str))
plt.show()
```
